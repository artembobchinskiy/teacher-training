## Описание проекта

**Проект для интернет-магазина "В один клик"**

**Описание проекта:** Интернет-магазин «В один клик» продаёт товары для детей, для дома, мелкую бытовую технику, косметику, продукты. Отчёт магазина за прошлый период показал, что активность покупателей начала снижаться. Привлекать новых клиентов не эффективно, так как о магазине знает бОльшая часть целевой аудитории. Возможный выход — удерживать активность постоянных клиентов. Сделать это можно с помощью персонализированных предложений.

**Цель исследования:** Разработать решение, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность.

**Задачи исследования:**
1.	Промаркировать уровень финансовой активности постоянных покупателей:
- «снизилась» - клиент стал покупать меньше товаров
- «прежний уровень» - клиент покупает прежнее количество товаров
2.	Собрать данные по клиентам по следующим группам признаков: 
-	**Коммуникация (сотрудников) с клиентом:** тип сервиса, резрашение на сообщения, маркетинг 6 месяцев, маркетинг текущего месяца, длительность истории с клиентом.
-	**Продуктовое поведение (покупателя):** наиболее частая категория продукта в заказе, среднее число просматриваемых категорий продуктов за визит, число неоплаченных продуктов в корзине за последние 3 месяца.
-	**Поведение (покупателя) на сайте:** время на сайте за прошлый год, время на сайте за предыдущий месяц, время на сайте за текущий месяц, минут за визит, число страниц за визит, сбои сайта.
-	**Финансовое поведение (покупателя):** разница в выручке предыдущего и предпредыдущего месяца, разница в выручке текущего и предыдущего месяца, выручка в текущем месяце.
3. Построить модель машинного обучения, которая предскажет вероятность снижения покупательской активности клиента в следующие три месяца.
4. Используя данные модели и данные о прибыльности клиентов, выделить сегмент покупателей и разработать для него персонализированное предложение для повышения покупательской активности.

**Ход исследования:** 
1. Подготовка данных.
2. Предобработка данных.
3. Исследовательский анализ данных.
4. Корреляционный анализ признаков в датасете.
5. Построение лучше модели прогнозирования.
6. Сегментация покупателей.
7. Анализ целевого сегмента.
8. Подведение итогов.

**Общий вывод:** разработана модель машинного обучения, позволяющая персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность.

**Описание данных**: Данные размещены в четырех датасетах: <br>
1. Файл "market_file.csv" содержит данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении и состоит из следующих столбцов: <br>
<br>
-	id — номер покупателя в корпоративной базе данных
-	Покупательская активность — рассчитанный класс покупательской активности (целевой признак)
-	Тип сервиса — уровень сервиса
-	Разрешить сообщать — информация о том, можно ли присылать покупателю дополнительные предложения о товаре; согласие на это даёт покупатель
-	Маркет_актив_6_мес — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев; это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента
-	Маркет_актив_тек_мес — количество маркетинговых коммуникаций в текущем месяце
-	Длительность — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте
-	Акционные_покупки — среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев
-	Популярная_категория — самая популярная категория товаров у покупателя за последние 6 месяцев
-	Средний_просмотр_категорий_за_визит — показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца
-	Неоплаченные_продукты_штук_квартал — общее число неоплаченных товаров в корзине за последние 3 месяца
-	Ошибка_сервиса — число сбоев, которые коснулись покупателя во время посещения сайта
-	Страниц_за_визит — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца
2. Файл "market_money.csv" содержит данные о выручке, которую получает магазин с покупателя (сколько покупатель всего потратил за период взаимодействия с сайтом) и состоит из следующих столбцов:
-	id — номер покупателя в корпоративной базе данных
-	Период — название периода, во время которого зафиксирована выручка
-	Выручка — сумма выручки за период
3. Файл "market_time.csv" содержит данные о времени (в минутах), которое покупатель провёл на сайте в течение периода и состоит из следующих столбцов:
-	id — номер покупателя в корпоративной базе данных
-	Период — название периода, во время которого зафиксировано общее время
-	минут — значение времени, проведённого на сайте (в минутах)
4. Файл "money.csv" содержит данные о среднемесячной прибыли покупателя за последние 3 месяца (какую прибыль получает магазин от продаж каждому покупателю) и состоит из следующих столбцов:
-	id — номер покупателя в корпоративной базе данных
-	Прибыль — значение прибыли

***Поскольку данные в датафрейме market_money по каждому пользователю разделены на несколько строк в соответствии с периодом, когда он осущестлвял покупки, преобразуем таблицу в формат, когда по пользователь будет описан только одной строкой. В этом случае необходимо столбец "период" разделить на три столбца, по каждому из которых будет просуммирована выручка:***

```python
# создадим новый датафрейм

market_money_new = market_money

# убедимся, что все столбцы имеют требуемые для преобразования типы данных (в противном случае преобразование вызовет ошибку)

market_money_new['id'] = market_money['id'].astype(str)
market_money_new['период'] = market_money['период'].astype(str)
market_money_new['выручка'] = market_money['выручка'].astype(float)

# создаем датафрейм, сгруппированный по 'id' и разделенный по категориям столбца 'период'

market_money_new = market_money_new.pivot_table(index='id', columns='период', values='выручка', aggfunc='sum', fill_value=0)

# сбросываем индекс

market_money_new = market_money_new.reset_index()

# удаляем мультииндекс названия столбцов

market_money_new.columns.name = None

# переименовываем столбцы для удобства

market_money_new.columns = ['id', 'выручка_предыдущий_месяц', 'выручка_препредыдущий_месяц', 'выручка_текущий_месяц']

market_money_new
```

*Для решения этой задачи подходит датафрейм market_money, поскольку в нем по столбцам "период" и "выручка" хранятся данные о выручке с каждого покупателя как раз за три месяца.*

*Необходимо найти всех покупаетелей (по столбцу id), у которых по столбцу "выручка" значения больше нуля сразу по трем категориям "текущий_месяц", "предыдущий_месяц" и "препредыдущий_месяц" столбца "период" датафрейма market_money.*

*Отметим, что данную задачу можно было понять по-другому: найти клиентов, которые были активны 3 месяца назад и более, но данный вариант отметается, так как нет данных об активности за 4, 5 и так далее месяца.*

```python
# суммируем выручку по пользователям и периодам в отдельном датафрейме

market_money_grouped = market_money.groupby(['id', 'период']).agg({'выручка': 'sum'}).reset_index()

# фильтрация записей из нового датафрейма, где выручка больше нуля, и формирование на его основе
# датафрейма, где по каждому пользователю указано количество пероиодов с ненулевой выручкой

market_money_grouped_num_per = market_money_grouped[market_money_grouped['выручка'] > 0].groupby('id')['период'].nunique()

# сохранение в датафрейм только тех пользователей, у которых по всем трем периодом выручка больше нуля

market_money_kriteria = market_money_grouped_num_per[market_money_grouped_num_per == 3].index

# фильтруем исходный датафрейм по найденным id, где выручка больше нуля по всем трем периодам

market_money_filtered = market_money[market_money['id'].isin(market_money_kriteria)]

market_money_filtered
```

## Использование пайплайнов

Перед созданием пайплайнов осуществим предварительную подготовку:

```python
# оценка, есть ли дисбаланс классов целевого признака

# повторное построение гистограммы по целевому признаку

data['покупательская_активность'].hist(bins=2, figsize=(10,5))
plt.title('Распределение покупателей по категориям покупательской активности')
plt.xlabel('Категории покупательской активности')
plt.ylabel('Покупатели, чел.');
```

Покупателей с прежним уровнем активности в 1.6 раза больше, чем покупателей со снизившимся уровнем активности. Налицо дисбаланс классов. В этой связи введем при разделении выборки на тренировочную и тестовую параметр stratify=y.

```python
# разделение данных на входные признаки (X) и целевой признак (y)

X = data.drop(['покупательская_активность'], axis=1)
y = data['покупательская_активность']
```

```python
# разделение датасета на обучающую и тестовую выборки
# учитывается НЕсбалансированность stratify=y, фиксируется пропорция test_size

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y
)
```

```python
# преобразование целевого категориального признака с использованием LabelEncoder
# в бинарный, где имеет значение порядок (прежний уровень активности лучше снизившегося)

encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
# вывод закодированных категорий целевого признака

for original, encoded in zip(encoder.classes_, range(len(encoder.classes_))):
    print(f"{original}: {encoded}")
```

```python
# создание списков с количественными и категориальными признаками

cat_col_names = X_train.select_dtypes(exclude='number').columns.tolist()
cat_col_names.remove('тип_сервиса')
ord_col_names = ['тип_сервиса']
num_col_names = X_train.select_dtypes(include='number').drop(columns='id').columns.tolist()
```

Создание пайплайна обработки категориальных данных OneHotEncoder:

```python
ohe_pipe = Pipeline([
    ('ohe', OneHotEncoder(drop='first', sparse=False, handle_unknown='ignore'))
])
```

```python
# пайплайн для Ordinal Encoding
    
ord_pipe = Pipeline([
    ('ordinal', OrdinalEncoder(categories=[['стандарт', 'премиум']], 
                               handle_unknown='use_encoded_value', 
                               unknown_value=np.nan))
])
```

Создание пайплайна подготовки данных (в котором количественные данные обрабатываются StandardScaler):

```python
# ColumnTransformer для обработки признаков с использованием StandardScaler

data_preprocessor_standard = ColumnTransformer(
    transformers=[
        ('cat', ohe_pipe, cat_col_names),
        ('num', StandardScaler(), num_col_names),
        ('ord', ord_pipe, ord_col_names)
    ],
    remainder='passthrough'  # передача остальных столбцов без изменений
)
```

Создание пайплайна подготовки данных (в котором количественные данные обрабатываются MinMaxScaler):

```python
# ColumnTransformer для обработки признаков с использованием MinMaxScaler

data_preprocessor_minmax = ColumnTransformer(
    transformers=[
        ('cat', ohe_pipe, cat_col_names),
        ('num', MinMaxScaler(), num_col_names),
        ('ord', ord_pipe, ord_col_names)
    ],
    remainder='passthrough'
)
```

```python
# создание общего пайплайна (объединение предварительной обработки и обучения модели)

pipe_final = Pipeline([
    ('preprocessor', data_preprocessor_standard),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

```python
# определение сетки параметров для поиска лучшей модели

param_grid = [
    # Словарь для модели DecisionTreeClassifier
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(3, 21),
        'models__max_features': range(2, 5),
        'models__min_samples_split': range(2, 21),
        'models__min_samples_leaf': range(1, 21),
        'preprocessor': [data_preprocessor_standard, data_preprocessor_minmax]  
    },
    
    # Словарь для модели KNeighborsClassifier
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 20),
        'models__p': [1, 2],
        'models__weights' : ['uniform', 'distance'],
        'preprocessor': [data_preprocessor_standard, data_preprocessor_minmax]   
    }
]
```

Поскольку выше достаточно большое количество параментов, то используем RandomizedSearchCV для поиска их лучшей комбинации. Не используем GridSearchCV, так как сетка требует больших вычислительных ресурсов на большом количестве параметров. Не используем OptunaSearchCV, так как не ожидается значительных нелинейностей в задаче, а также этот метод требует бОльших вычислительных ресурсов.

```python
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_distributions=param_grid,
    n_iter=200,
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search.fit(X_train, y_train)
```

```python
# выберем лучшую модель

print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
```

Лучшей моделью оказался KNeighborsClassifier с параметрами: 8 ближайших соседей, метрикой расстояния L1-норма (манхэттенское расстояние), весом каждого соседа обратно пропорциональным расстоянию до предсказываемой точки.

```python
# проверим работу модели метрикой ROC-AUC

print ('Метрика ROC-AUC лучшей модели на кросс-валидации:', randomized_search.best_score_)

# извлечение и обучение лучшей модели

best_model = randomized_search.best_estimator_.named_steps['models']

# применение предварительной обработки данных

X_train_processed = randomized_search.best_estimator_.named_steps['preprocessor'].transform(X_train)
X_test_processed = randomized_search.best_estimator_.named_steps['preprocessor'].transform(X_test)

# Обучение модели на предварительно обработанных данных
best_model.fit(X_train_processed, y_train)

# Предсказание на тренировочной выборке
y_train_pred_proba = best_model.predict_proba(X_train_processed)[:, 1]
print(f'Метрика ROC-AUC лучшей модели на тренировочной выборке: {roc_auc_score(y_train, y_train_pred_proba)}')

# Предсказание на тестовой выборке
y_test_pred_proba = best_model.predict_proba(X_test_processed)[:, 1]
print(f'Метрика ROC-AUC лучшей модели на тестовой выборке: {roc_auc_score(y_test, y_test_pred_proba)}')
```

Эта модель почти в 99.99% случаев верно предсказала целевые признаки тестового набора данных.

**Метрика ROC-AUC была выбрана, поскольку она оценивает модели по всем порогам вероятности - более обобщенно, чем, например, F1-Score, подходит для задач бинарной классификации (задача проекта именно такая), устойчива к несбалансированным данным (некоторые из признаков несбалансированны).
Теоретически можно было вместо ROC-AUC взять Precision и Recall, но они не дают общей картины, а Accuracy не подходит к несбалансированным данным.**

**Промежуточные выводы:**
1. Был выявлен дисбаланс между классами в целевом признаке: покупателей с прежним уровнем активности в 1.6 раза больше, чем покупателей со снизившимся уровнем активности. Это было учтено в модели машинного обучения.

4.	Метрика лучшей модели на тренировочной выборке: 0.9999776636140272. Метрика ROC-AUC на тестовой выборке: 0.9975124378109453 Эта модель почти в 99.75% случаев верно предсказала целевые признаки тестового набора данных.

Используем результаты полученной модели прогнозирования покупательской активности для выделения тех покупателей, риск снижения покупательской активности которых высокий, чтобы бизнес мог предпринять меры по работе с ними. Например, выделим в отдельный сегмент тех покупателей, по которым модель предсказывает вероятность снижения активности выше 0.7, такие покупатели могут считаться "наиболее подверженными риску снижения покупательской активности".

Предлагается особое внимание признакам уделить при оценке покупателей из сегмента "покупательская активность снизилась".
21. Результаты полученной модели прогнозирования снижения покупательской активности клиентов сервиса позволили сегментировать всех покупателей на две группы: тех, у кого вероятность отнесения к покупателям с снижающейся активностью составила от 0.7 до 1, то есть повышенная; и тех, у кого таковая вероятность от 0 до 0.7, то есть у кого низкий риск отнесения к покупателям со снижающейся активностью.
22.	В качестве целевого признака для оценки указанных сегментов была взята прибыль, которую компания зарабатывает с каждого клиента.
23.	Количество пользователей, у которых высокая вероятность снижения покупательской активности (от 0.7 до 1) составило 464 чел. С данными покупателями необходимо проведение мероприятий по повышению покупательской активности.
24.	Покупатели, у которых высокая вероятность снижения покупательской активности приносят меньше прибыли, нежели те, у кого активность сохраняется. Тем не менее, прибыль их является значительной. Распределена прибыль по покупателям обоих типов активности нормально с приблизительно одинаковым средним.
25.	Была повторно проведена оценка корреляций между признаками и типами активности (после выставления типов в соответствии с решением модели машинного обучения). Все корреляции, о которых говорилось в предыдущем выводе, выросли (повысилась их чувствительность).
26.	Пользователи с высокой вероятностью снижения покупательской активности в премиум сегменте приносят приблизительно в 2 раза меньше прибыли, чем в стандартном сегменте, при этом пользователи с сохраняющейся активностью приносят в целом больше прибыли в премиум и стандарт сегментах с сохранением приблизительно такого же соотношения 1 к 2.
27.	Те пользователи, которые разрешили сообщать о дополнительных предложениях о товарах, приносят больше прибыли, чем те, которые это запретили. Связь с типом активности пользователей на данном графике не наблюдается.
28.	Пользователи, с которыми было 4 маркетинговых коммуникаций в текущем месяце принесли больше прибыли, чем пользователи, с которыми таковых коммуникаций было 3 или 5. Маркетинговые коммуникации в текущем месяце не связаны с типом активности покупателей.
29.	Там, где среднее количество рекламы за последние 6 месяцев от 1 до 3 шт., в основном сконцентрированы пользователи, у которых высока вероятность снижения покупательской активности, при этом прибыль может быть разной. Это говорит о том, что следует для таких пользователей повышать количество маркетинговых коммуникаций.
30.	Пользователи с высоким риском снижения активности проводят меньше минут на сайте, чем большинство пользователей.
31.	Зависимость между прибылью и количеством дней после регистрации на сайте в разрезе типов активности – отсутствует.
32.	Среди покупателей, как ранее уже было выяснено, есть два сегмента: те, у кого среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев составляет от 0 до 40%, и те, у кого доля составляет от 80 до 100%. Заметим, что в сегменте доли от 80 до 100% в основном находятся покупатели, у которых высока вероятность снижения покупательской активности.
33.	Больше всего прибыли приносят товары для детей - порядка 1300, следующей по пробыли категорией является домашний текстиль - 1000, затем косметика и аксесуары - 900, а остальные категории еще меньше. Внутри категорий разное соотношение покупателей с разным типом активности. Если в категории "Мелкая бытовая техника и электроника" покупателей со снижающейся активностью крайне мало, то в "Товарах для детей" и "Косметика и аксесуары" их почти половина от всех, также их достаточно много в "Домашний текстиль и "Кухонная посуда".
34.	Покупатели с высокой вероятностью снижения покупательской активности в основном просматривают 2, 3 или 1 категории, и очень редко 6, 5 или 4 категории. Вероятно, они всегда ориентированы на конкретные категории, например, которые дешевле стоят.
35.	Покупатели с высокой вероятностью снижения покупательской активности в равной пропорции с покупателями с сохраняющимся уровнем активности имеют 1-4 неоплаченных товаров в корзине, но если говорить про покупателей, у которых в корзине 6-10 неоплаченных товаров, то почти всегда это покупатели с высокой вероятностью снижения покупательской активности. Для 5 товаров в корзине равное количество и тех и тех покупателей, а 0 товаров в корзине в основном имеют те, у кого покупательская активность сохраняется.
36.	Покупатели с высокой вероятностью снижения покупательской активности в основном просматривают за один визит от 1 до 7 страниц. В то же время покупатели с высокой вероятностью сохранения покупательской активностью просматривают больше страниц: в основном от 4 до 17.
37.	Для покупателей, у которых высокая вероятность отнесения к тем, у которого покупательская активность снизилась, можно предложить маркетинговые акции, направленные на поддержку покупательской активности (скидки, промокоды, купоны, кэшбек) в тех категориях товаров, которые ранее этими пользователями обходились стороной.
